```python
import pandas as pd
import numpy as np
import math
from tqdm import tqdm
from pathlib2 import Path
```

```python
file_dir = Path('/Users/zhangjunjie/Downloads/FATDataset')
test_info_path = file_dir / 'test_set.txt'
train_info_path = file_dir / 'train_set.txt'
```

```python
def convert_table(raw_info):
    num_crops = raw_info.shape[0]
    idx = 0
    new_info = pd.DataFrame()
    new_row = raw_info.iloc[idx]
    new_row.iloc[2] /= new_row.iloc[1]
    new_row.drop([1])
    pic_name = new_row.iloc[0]
    pbar = tqdm(total=num_crops)
    while idx < (num_crops - 1):
        idx += 1
        print(idx)
        current_row = raw_info.iloc[idx]
        current_pic_name = current_row.iloc[0]
        current_row.iloc[2] /= current_row.iloc[1]
        if current_pic_name == pic_name:
            new_row = new_row.append(current_row[2:],ignore_index=True)
        else:
            new_info = new_info.append(new_row, ignore_index=True)
            pic_name = current_pic_name
            new_row = raw_info.iloc[idx]
            new_row.iloc[2] /= new_row.iloc[1]
            new_row.drop([1])
        
        pbar.update(1)
    pbar.close()
    
    return new_info

        
```

```python
train_info = pd.read_table(train_info_path, sep=(' '), header=None)
new_train_info = convert_table(train_info)
```
